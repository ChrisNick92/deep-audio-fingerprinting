```python
import os
project_path = os.path.dirname(os.path.abspath(os.getcwd()))
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def dsigmoid(x):
    return sigmoid(x)*(1 - sigmoid(x))
```

```python
v_sig = np.vectorize(sigmoid)
v_dsig = np.vectorize(dsigmoid)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
x = np.linspace(start=-8, stop=8, num=200)
ax.plot(x, v_sig(x), label=r'$S(x) = \frac{1}{1+e^{-x}}$')
ax.plot(x, v_dsig(x), label=r"$S'(x) = \frac{e^{-x}}{(1+e^{-x})^2}$")

plt.legend()
plt.savefig(os.path.join(project_path, 'images/sigmoid.pdf'), format='pdf')
plt.show()
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
tanh_vec = np.vectorize(tanh)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(start=-6, stop=6, num=200)
ax.plot(x, tanh_vec(x), label=r"$\frac{e^x-e^{-x}}{e^x+e^{-x}}$")
plt.legend()
# plt.savefig(os.path.join(project_path, 'images/tahn.pdf'), format='pdf')
plt.show()
```
